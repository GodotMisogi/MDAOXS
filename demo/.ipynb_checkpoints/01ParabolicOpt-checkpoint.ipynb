# Optimizing a Paraboloid â€“ The TL;DR Version [link](http://openmdao.org/twodocs/versions/latest/examples/tldr_paraboloid.html)

we have the following problem

$\min{f(x,y)= (x-3)^2 + x*y + (y+4)^2 - 3}$

such that

$-50\leq x\leq 50$

$-50\leq y\leq 50$

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def f(x,y):
    return (x-3)**2 + x*y + (y+4)**2 - 3
x = np.arange(-50,50,0.1)
y = np.arange(-50,50,0.1)
[X,Y] = np.meshgrid(x,y)
f(X,Y)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,f(X,Y),cmap=cm.coolwarm, linewidth=0, antialiased=False)
plt.show()
```

```python
np.min(f(X,Y))
```

## In OpenMDAO

```python
from openmdao.api import Problem, ScipyOptimizer, ExecComp, IndepVarComp

# build the model
prob = Problem()
indeps = prob.model.add_subsystem('indeps', IndepVarComp())
indeps.add_output('x', 3.0)
indeps.add_output('y', -4.0)

prob.model.add_subsystem('paraboloid', ExecComp('f = (x-3)**2 + x*y + (y+4)**2 - 3'))

prob.model.connect('indeps.x', 'paraboloid.x')
prob.model.connect('indeps.y', 'paraboloid.y')

# setup the optimization
prob.driver = ScipyOptimizer()
prob.driver.options['optimizer'] = 'SLSQP'

prob.model.add_design_var('indeps.x', lower=-50, upper=50)
prob.model.add_design_var('indeps.y', lower=-50, upper=50)
prob.model.add_objective('paraboloid.f')

prob.setup()
prob.run_driver()
```

# Construct a ExplicitComp

```python
from __future__ import division
import numpy as np
from openmdao.api import ExplicitComponent

class ParaboloidComp(ExplicitComponent):
    def initialize(self):
        pass
    
    def setup(self):
        self.add_input('x')
        self.add_input('y')
        self.add_output('f')
        self.declare_partials('*', '*')

    def compute(self, inputs, outputs):
        outputs['f'] = (inputs['x']-3)**2 + inputs['x']*inputs['y'] + (inputs['y']+4)**2 - 3

    def compute_partials(self, inputs, partials):
        partials['f', 'x'] = 2*(inputs['x']-3) + inputs['y']
        partials['f', 'y'] = inputs['x'] + 2*(inputs['y']+4)
```

## Form a Group 

```python
from __future__ import division
import numpy as np

from openmdao.api import Group, IndepVarComp
class ParaboloidGroup(Group):
    def initialize(self):
        pass

    def setup(self):
        inputs_comp = IndepVarComp()
        inputs_comp.add_output('x',1)
        inputs_comp.add_output('y',1)
        self.add_subsystem('inputs_comp', inputs_comp)
        
        paraboloid_comp = ParaboloidComp()
        self.add_subsystem('paraboloid_comp',paraboloid_comp)
        
        self.connect('inputs_comp.x', 'paraboloid_comp.x')
        self.connect('inputs_comp.y', 'paraboloid_comp.y')
        
        self.add_design_var('inputs_comp.x', lower=-50, upper=50.)
        self.add_design_var('inputs_comp.y', lower=-50, upper=50.)
        self.add_objective('paraboloid_comp.f')
```

```python
import numpy as np

# for performance profiling
from openmdao.devtools import iprofile as tool
from openmdao.api import Problem, ScipyOptimizer

method = [ ('*compute_partials*', (ParaboloidComp, ))]
tool.setup(methods=method)
tool.start()

prob = Problem(model=ParaboloidGroup())

prob.driver = ScipyOptimizer()
prob.driver.options['optimizer'] = 'SLSQP'
prob.driver.options['tol'] = 1e-9
prob.driver.options['disp'] = True
prob.setup(force_alloc_complex=True) 
prob.check_partials(method='cs')
 
prob.check_partials(method='fd')
prob.run_driver()
print(prob['paraboloid_comp.f'])
tool.stop()
```
